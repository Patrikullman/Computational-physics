```python
import numpy as np
import matplotlib.pyplot as plt


array = np.genfromtxt('lattice_para_vs_Ep.csv', delimiter=',')

#fig, ax = plt.subplots()
plt.plot(array[:, 0]**3, array[:, 1], 'o', label = 'Calculated pot. energy')

coeff = np.polyfit(array[:, 0]**3, array[:, 1], 2)
x = np.linspace(61, 69, 100)
y = coeff[2] + coeff[1]*x + coeff[0]*x**2
print(coeff)
print("Minimum energy volume: ", -coeff[1]/(2*coeff[0]))
print("Optimal lattice para: ", (-coeff[1]/(2*coeff[0]))**(1/3))
plt.plot(x,y,'--', label = 'Quadratic fit')
plt.title('Energy-Volume diagram of aluminium', fontsize = 14)
plt.xlabel('Volume ($^3$)', fontsize = 12) 
plt.ylabel('Energy (eV/unit cell)', fontsize = 12)

#ax.set_xlabel('time (arb.unit)')
#ax.set_ylabel('signal (arb.unit)')
#ax.grid()
plt.legend(loc = 1, prop={'size':12})
plt.savefig("pot_energy_vs_unit_cell.png", bbox_inches='tight')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

array = np.genfromtxt('dist_between_points.csv', delimiter=',')

pos_array = np.zeros((256,256))
dist_sum = np.zeros((256))

for i in range(256):
    for j in range(256):
        pos_array[i,j] = array[i*256+j,1]
#print(pos_array[0,:])

for i in range(256):
    for j in range(256):
        dist_sum[i] += pos_array[i,j]
print(np.max(dist_sum/256))
        
plt.hist(dist_sum/256, bins = 30)
plt.show()
        
counts, bin_edges = np.histogram(dist_sum/256, 30)
print(counts)
#print(bin_edges)

delta_r = np.zeros((30))
for i in range(0,30):
    delta_r[i] = bin_edges[i+1]-bin_edges[i]
    
print(delta_r)
N_ideal = np.zeros((30))
for i in range(0,30):
    N_ideal[i] = (255*4*np.pi / (3*64)) * (3*i**2 - 3*i + 1)*(delta_r[i])**3
    
#print(N_ideal)

gr = np.zeros((30))
#sum_count = 0
for i in range(0,30):
    #sum_count += counts[i]
    gr[i] = counts[i]/N_ideal[i]
    
r_mean = np.zeros((30))
for i in range(30):
    r_mean[i] = bin_edges[i+1]
print(r_mean)
print(gr.shape)

plt.plot(r_mean, gr)
plt.show()

print(N_ideal)
print(counts)


x = np.linspace(0,23,30)
plt.plot(x, counts)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

array = np.genfromtxt('dist_between_points.csv', delimiter=',')

array_sorted = np.sort(array[:,1])
array = np.trim_zeros(array_sorted, trim = 'f')



           
plt.hist(array, bins = 50)
plt.show()
        
counts, bin_edges = np.histogram(array, 50)


print(counts)
#print(bin_edges)

delta_r = np.zeros((50))
for i in range(0,50):
    delta_r[i] = bin_edges[i+1]-bin_edges[i]
    
#print(delta_r)
N_ideal = np.zeros((50))
for i in range(0,50):
    N_ideal[i] = (255*4*np.pi / (3*64)) * (3*i**2 - 3*i + 1)*(delta_r[i])**3
    
print(N_ideal)

gr = np.zeros((50))

for i in range(0,50):
    gr[i] = counts[i]/N_ideal[i]
    
r_mean = np.zeros((50))
for i in range(50):
    r_mean[i] = bin_edges[i+1]


plt.plot(r_mean, gr)
plt.axvline(x=1.5, color = 'r')
plt.axhline(y = 1, color = 'magenta')
plt.show()





```
